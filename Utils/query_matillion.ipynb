```python
import pandas as pd
import json
from datetime import datetime
```

```python
#path = 'export_interseguro_data.json'
path = 'export (4).json'
```

```python
with open(path, 'r', encoding='utf-8') as archivo:
    datos_json = json.load(archivo)
```

```python
metadada_jobs_created = datos_json['orchestrationJobs']
jobs_data = [(job['id'], datetime.utcfromtimestamp(job['created'] / 1000)) for job in metadada_jobs_created]
df = pd.DataFrame(jobs_data, columns=['ID', 'FECHA_CREACION'])    
```

```python
with open('archivo.json', 'w') as file:
    json.dump(datos_json['orchestrationJobs'][0]['components'], file)
```

```python
datos_json['orchestrationJobs'][0]['components']
```

```python
# Listas para almacenar los datos
job_names = []
job_ids = []
job_files = []

data = datos_json['jobsTree']['children'][1]['children']

for i, job in enumerate(data):
    #print(f"Nombre Carpeta: {job['name']}")
    job_file = job['name']
    #print(job_file)
    temp = datos_json['jobsTree']['children'][1]['children'][i]['jobs']
    for j, jobs_info in enumerate(temp):
        job_id = jobs_info['id']
        job_name = jobs_info['name']
        
        #print(f"Nombre Job: {jobs_info['id']},{jobs_info['name']}")
        
        job_names.append(job_name)
        job_ids.append(job_id)
        job_files.append(job_file)
```

```python
df_data = pd.DataFrame({
    'ID': job_ids,
    'NOMBRE_FILE': job_files,
    'NOMBRE_JOB': job_names
})
```

```python
merged_df = pd.merge(df_data, df, on='ID', suffixes=('_left', '_right'), how='inner')
```
